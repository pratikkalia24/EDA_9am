```python
basic = eval(input("Enter Ramesh's basic salary: "))

da = 0.40 * basic   
hra = 0.20 * basic  

gross = basic + da + hra

print("Gross Salary =",gross)
```

```python
km = eval(input("Enter distance in kilometers: "))

meters = km * 1000
feet = km * 3280.84
inches = km * 39370.1
cm = km * 100000

print("Meters =", meters)
print("Feet =", feet)
print("Inches =", inches)
print("Centimeters =", cm)
```

```python
m1 = eval(input("Enter marks of subject 1: "))
m2 = eval(input("Enter marks of subject 2: "))
m3 = eval(input("Enter marks of subject 3: "))
m4 = eval(input("Enter marks of subject 4: "))
m5 = eval(input("Enter marks of subject 5: "))

aggregate = m1 + m2 + m3 + m4 + m5
percentage = (aggregate / 500) * 100

print("Aggregate Marks =", aggregate)
print("Percentage =", percentage, "%")
```

```python
f = eval(input("Enter temperature in Fahrenheit: "))

c = (f - 32) * 5/9

print("Temperature in Celsius =", c)
```

```python
length = eval(input("Enter length of rectangle: "))
breadth = eval(input("Enter breadth of rectangle: "))
radius = eval(input("Enter radius of circle: "))

area_rect = length * breadth
peri_rect = 2 * (length + breadth)

area_circle = 3.14 * radius * radius
circum_circle = 2 * 3.14 * radius

print("Area of Rectangle =", area_rect)
print("Perimeter of Rectangle =", peri_rect)
print("Area of Circle =", area_circle)
print("Circumference of Circle =", circum_circle)
```

```python
C = input("Enter value for C: ")
D = input("Enter value for D: ")

C, D = D, C

print("After swapping:")
print("C =", C)
print("D =", D)
```

```python
num = eval(input("Enter a five-digit number: "))

rev = eval(str(num)[::-1])

print("Reversed Number =", rev)
```

```python
num = input("Enter a four-digit number: ")

first = eval(num[0])
last = eval(num[-1])

print("Sum of first and last digit =", first + last)
```

```python
population = 80000

men = 0.52 * population
women = population - men

literate = 0.48 * population
literate_men = 0.35 * population
literate_women = literate - literate_men

illiterate_men = men - literate_men
illiterate_women = women - literate_women

print("Illiterate Men =", illiterate_men)
print("Illiterate Women =", illiterate_women)
```

```python
amount_100s = eval(input("Enter the amount in hundreds: "))

amount = amount_100s * 100

notes_100 = amount // 100
remaining = amount % 100

notes_50 = remaining // 50
remaining = remaining % 50

notes_10 = remaining // 10

print("100 Rs notes =", notes_100)
print("50 Rs notes =", notes_50)
print("10 Rs notes =", notes_10)
```
