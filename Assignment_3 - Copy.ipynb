```python
# 1
n1=eval(input('enter the number1:'))
n2=eval(input('enter the number2:'))
n3=eval(input('enter the number3:'))
avg=(n1+n2+n3)/3
round_avg=round(avg)
print(f'the avg of there number is :{round_avg}')
```

```python
# 2
n1=eval(input('enter the number1:'))
n2=eval(input('enter the number2:'))
sub=(n1-n2)
mul=(n1*n2)
div=(n1/n2)
round_sub=round(sub) 
round_mul=round(mul) 
round_div=round(div)
print(f'the sub of two number is :{round_sub}')
print(f'the mul of two number is :{round_mul}')
print(f'the div of two number is :{round_div}')
```

```python
# 3
pi=eval(input('enter the number:'))
r=eval(input('enter the number:'))
area_of_circle=pi*r*r
round_area_of_circle=round(area_of_circle)
print(f'the area of circle is :{round_area_of_circle}')
```

```python
# 4
h=eval(input('enter the number:'))
b=eval(input('enter the number:'))
area_of_triangle=0.5*b*h
round_area_of_triangle=round(area_of_triangle)
print(f'the area of triangle is :{round_area_of_triangle}')
```

```python
# 5
bill_amount=eval(input('enter the number:'))
tip_amount=eval(input('enter the number:'))
total_amount=bill_amount+tip_amount
round_total_amount=round(total_amount)
print(f'the total amount is :{round_total_amount}')
```
