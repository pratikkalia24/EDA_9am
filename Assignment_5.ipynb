```python
a = eval(input("Enter first number: "))
b = eval(input("Enter second number: "))

print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```

```python
import math

r = eval(input("Enter radius of circle: "))
area = math.pi * r * r

print("Area of circle =", area)
```

```python
h = eval(input("Enter height of triangle: "))
b = eval(input("Enter base of triangle: "))

area_triangle = 0.5 * b * h
print("Area of triangle =", area_triangle)
```

```python
a = eval(input("Enter number 1: "))
b = eval(input("Enter number 2: "))
c = eval(input("Enter number 3: "))

avg = (a + b + c) / 3
print("Average =", avg)
```

```python
bill = eval(input("Enter bill amount: "))
tip = eval(input("Enter tip amount: "))

total = bill + tip
print("Total bill =", total)
```

```python
salary = eval(input("Enter salary: "))
tax_percent = eval(input("Enter tax percentage: "))

tax_amount = (salary * tax_percent) / 100
print("Tax amount =", tax_amount)
```

```python
dollar = eval(input("Enter amount in dollars: "))

rupees = dollar * 80
print("Amount in rupees =", rupees)
```

```python
kg = eval(input("Enter weight in kilograms: "))

pounds = kg * 2.2
print("Weight in pounds =", pounds)
```
